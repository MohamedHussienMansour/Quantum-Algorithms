```python
from qiskit import *
```

```python
qiskit.__version__
```

```python
circuit = QuantumCircuit(3,3)
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1)
circuit.cx(1,2)
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)

circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1],[1,0])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cx(0,2)
circuit.draw(output='mpl')
```

```python
circuit.measure(2,2)
simulator=Aer.get_backend("qasm_simulator")
result=execute(circuit,backend=simulator,shots=1024)
```

```python

# Transpile the circuit for the simulator backend
transpiled_circuit = transpile(circuit, backend=simulator)

# Run the transpiled circuit on the simulator
result = simulator.run(transpiled_circuit, shots=1024).result()
```

```python
counts=result.get_counts()
```

```python
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts,figsize=(3,3))
```
